```python
def calculas_frequências(df, coluna1, coluna2):
    resultados = {}
    for col in coluna1:  # coluna1 = variáveis ordinais

        # Calcular frequências
        abs_freq = df[col].value_counts().sort_index()
        abs_freq_acum = np.cumsum(abs_freq)
        rel_freq = df[col].value_counts(normalize=True).sort_index() * 100
        rel_freq_acum = np.cumsum(rel_freq)
        
        resultados[col] = pd.DataFrame({
            'Frequência absoluta': abs_freq,
            'Frequência absoluta acumulada': abs_freq_acum,
            'Frequência relativa (%)': rel_freq,
            'Frequência relativa acumulada (%)': rel_freq_acum
        })
        print('\nVariáveis ordinais')
        print(resultados[col])
        
        # Criar gráfico de barras para variáveis ordinais
        plt.figure(figsize=(8, 4))
        sns.barplot(x=abs_freq.index, y=abs_freq.values, color='blue')
        plt.xlabel(col)
        plt.ylabel('Frequência Absoluta')
        plt.title(f'Gráfico de barras - {col}')
        plt.show()
    
    resultados = {}
    for col in coluna2:  # coluna2 = variáveis não ordinais

        # Calcular frequências
        abs_freq = df[col].value_counts().sort_index()
        rel_freq = df[col].value_counts(normalize=True).sort_index() * 100
        
        resultados[col] = pd.DataFrame({
            'Frequência absoluta': abs_freq,
            'Frequência relativa (%)': rel_freq,
        })
        print('\nVariáveis não ordinais')
        print(resultados[col])
        
        # Criar gráfico de barras para variáveis não ordinais
        plt.figure(figsize=(8, 4))
        sns.barplot(x=abs_freq.index, y=abs_freq.values, color='green')
        plt.xlabel(col)
        plt.ylabel('Frequência Absoluta')
        plt.title(f'Gráfico de barras - {col}')
        plt.show()

var_ordinais = ["CLASIFFICATION_FINAL"]
var_nao_ordinais = [
    "USMER", "MEDICAL_UNIT", "SEX", "PATIENT_TYPE", "INTUBED", "PNEUMONIA", "PREGNANT", "DIABETES", "COPD", "ASTHMA",
    "INMSUPR", "HIPERTENSION", "OTHER_DISEASE", "CARDIOVASCULAR", "OBESITY", "RENAL_CHRONIC", "TOBACCO", "ICU"
]

calculas_frequências(df, var_ordinais, var_nao_ordinais)
```

```python
def plot_crosstab_with_stats(df, var_independentes, var_dependentes, cmap="Blues"):

    for var in var_dependentes:
        
        contingency_table = pd.crosstab(df[var_independentes], df[var])

        chi2, p, dof, expected = chi2_contingency(contingency_table)

        print(f"Teste Qui-Quadrado {var_independentes} vs {var}: χ²={chi2:.4f}, p-valor={p:.4f}")

        # Plot heatmap
        plt.figure(figsize=(6, 4))
        sns.heatmap(contingency_table, annot=True, fmt="d", cmap=cmap, linewidths=0.5)
        plt.title(f"{var_independentes} vs {var}")
        plt.xlabel(var)
        plt.ylabel(var_independentes)
        plt.show()

# Definir lista de variáveis dependentes para comparar com MORTALITY
var_dep = [
    "USMER","MEDICAL_UNIT","SEX","PATIENT_TYPE","INTUBED","PNEUMONIA","PREGNANT","DIABETES",
    "COPD","ASTHMA","INMSUPR","HIPERTENSION","OTHER_DISEASE","CARDIOVASCULAR",
    "OBESITY","RENAL_CHRONIC","TOBACCO","CLASIFFICATION_FINAL","ICU"
]

# Chamar a função
plot_crosstab_with_stats(df, var_independentes="MORTALITY", var_dependentes=var_dep)
```

```python
# Calcula os p-valores para cada variável (menos a 'MORTALITY') usando teste qui-quadrado
variables_to_test = df.columns.difference(['MORTALITY'])

# Lista para armazenar os p-valores
p_values_results = []

for var in variables_to_test:
    contingency_table = pd.crosstab(df[var], df['MORTALITY'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    p_values_results.append({'Variable': var, 'p-value': p})  # Corrected dictionary keys

# Converter resultados para DataFrame
p_values_df = pd.DataFrame(p_values_results)

# Printar cada variável com o respectivo p-valor
for index, row in p_values_df.iterrows():
    print(f"\nVariável: {row['Variable']}, p-value: {row['p-value']}")

# Identificar as colunas para remoção (p-valor > 0.05)
columns_to_drop = p_values_df[p_values_df['p-value'] > 0.05]['Variable'].tolist()

# Drop das colunas
df2 = df.drop(columns=columns_to_drop)

print("\nColunas removidas:", columns_to_drop)
```

As variáveis foram avaliadas pelo teste Qui-Quadrado e as colunas com p-value acima da significância, que consideramos ser 5%, foram removidas.

```python
def remover_colunas_alto_vif(dataframe, limite_vif=10):

    vif = pd.DataFrame()
    vif['VIF'] = [variance_inflation_factor(dataframe.values, i) for i in range(dataframe.shape[1])]
    vif['variavel'] = dataframe.columns

    # Identificar colunas para exclusão
    colunas_para_excluir = vif[vif['VIF'] > limite_vif]['variavel']
    print(colunas_para_excluir)
    X2 = dataframe.drop(columns=colunas_para_excluir, axis=1)

    return X2

# Usar a função para criar X2
X2 = remover_colunas_alto_vif(X)
```

```python
# Pressuposto 5
# Teste de Hosmer-Lemeshow


p_estimate = modelo2.predict(sm.add_constant(train_X2))  # Ensure constant column is added

# Creating a DataFrame for observed and expected events
hl_df = pd.DataFrame({
    "P_i": p_estimate,
    "Outcome": train_Y
})

# Dividing into deciles based on predicted probabilities
hl_df["decile"] = pd.qcut(hl_df["P_i"], 10)  # Split into 10 groups

# Summing observed and expected events grouped by deciles
obsevents_1 = hl_df.groupby("decile")["Outcome"].sum()
obsevents_0 = hl_df.groupby("decile")["Outcome"].count() - obsevents_1
expevents_1 = hl_df.groupby("decile")["P_i"].sum()
expevents_0 = hl_df.groupby("decile")["P_i"].count() - expevents_1

# Hosmer-Lemeshow test statistic
hl = (((obsevents_0 - expevents_0) ** 2) / expevents_0).sum() + \
     (((obsevents_1 - expevents_1) ** 2) / expevents_1).sum()

# Degrees of freedom = number of groups (10) minus 2
df = 10 - 2

# Compute the p-value for the chi-squared distribution
pvalue = 1 - chi2.cdf(hl, df)

# Output the result
print(f"Hosmer-Lemeshow Test p-value: {pvalue}")
```

Como o p-value do teste de Hosmer-Lemeshow é menor que a significância considerada, conclui-se que, este modelo não é adequado.
Consequentemente o pressuposto 5 não se verifica.

```python
test_X2 = sm.add_constant(test_X2)

p_estimate = modelo2.predict(test_X2) 
prediction = np.round(p_estimate)  # cut-off=0.5 <=> prediction = np.where(p_estimate >= 0.5, 1, 0)
cm = confusion_matrix(test_Y, prediction)
print(cm)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
# [row, column]
TP = cm[1, 1]
TN = cm[0, 0]
FP = cm[0, 1]
FN = cm[1, 0]

print("Acurácia: %.4f" % ((TP+TN) / float(TP + FN+TN+FP))) #proporção de previsões corretas feitas pelo modelo em relação ao total de previsões
print("Erro: %.4f" % ((FP+FN) / float(TP + FN+TN+FP))) #proporção de previsões erradas feitas pelo modelo(1-acurácia)
print("Sensibilidade: %.4f" % (TP / float(TP + FN))) #taxa deverdadeiros positivos
print("Especificidade: %.4f" % (TN / float(TN + FP))) #taxa de verdadeiros negativos
print("Precisão: %.4f" % (TP / float(TP + FP))) #proporção de verdadeiros positivos
print("Recall : %.4f" % (TP / float(TP + FN))) #taxa de verdadeiros negativos
print("F1 score  : %.4f" % (2*TP / float(2*TP + FP+FN)))
```
